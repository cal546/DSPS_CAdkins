Lets start by looking at a single chi sq distribution of 100 elements

```python
# Note: I am adding imports here, to show the workflow: 
# at this point I realize what packages I need and I import them. 
# but this is a PEP8 violation. 
# It is a good habit to put the includes at the top 
# so the code fails right away if the packages are missing. 
# Otherwise a user may have a piece of code running for a long time, 
# and then fail when the package cannot be imported... 
# (unless they are only imported under conditions, e.g. in an if statement)

import numpy as np
np.random.seed(456)
# prepare a container for all the distribution: 
# a dictionary so you can call each family of distribution by "key" 
md = {}
# mydists is an empty disctionary; 
# if you are not familiar with python dictionaries see https://www.pythonforbeginners.com/dictionary/how-to-use-dictionaries-in-python
md['chisq'] = {}
# mydists['chisq'] is an element of the mydist dictionary, itself a dictionary
md['chisq'][100] = np.random.chisquare(mymean, size=100)
# mydists now is a dictionary with one object in it:
#             a dictionary by the key 'chisq'
# mydists['chisq'] is a dictionary with one object in it: 
#             an array of 100 elements by the key 100
pl.hist(md['chisq'][100], bins = 30)
pl.ylabel('N')
pl.xlabel('x');
```

### Figure 1: Chi sq distribution of a sample of size 100 for a Chi square distribution with  100 degrees of freedom. 

## let numpy calculate what the mean and stdev of this distribution are

```python
print ("Chisq mean: %.2f, standard deviation: %.2f"%(md['chisq'][100].mean(), md['chisq'][100].std()))
```

## Now let's do it for 100 Chisq distributions
we want distributions of different sizes (all integers between 10 and 2000). It's a good idea to create more small distributions, cause, as we will see, their means will be more noisy (due to the law of large numbers which we discussed in class). Here are some tricks to create the distribution sizes 

```python
mysize = (2000 / (np.array(range(1, 100)))).astype(int)
# mysize = 1000 / (np.array(range(1, 100) + [10]))
# mysize = (np.random.rand(100) * 1000).astype(int)
print ("shape of the 'size' list that contains the size of each distribution", mysize.shape)
```

```python
pl.plot(mysize, '.')
pl.xlabel("index")
pl.ylabel("size of the distributions to be created");
```

```python

# fill dictionary in in a for loop. not pythonic, but easily readable
np.random.seed(45)
for n in mysize:
    md['chisq'][n] = np.random.chisquare(df, size = n) #all distributions have mean=df=100
    
# save the means for plotting later    

md['chisq']['means'] = {}
# prepare the figure
axchisq_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['chisq']:
    if not type(nn) == str: # needed because one element is "mean"
        md['chisq']['means'][nn] = md['chisq'][nn].mean()
        #and plot it
        axchisq_mu_n.plot(nn, md['chisq']['means'][nn], 'o')
        axchisq_mu_n.set_xlabel('sample size', fontsize=18)
        axchisq_mu_n.set_ylabel('sample mean', fontsize=18)
axchisq_mu_n.set_title('Chi squared', fontsize=18)
axchisq_mu_n.plot([min(mysize), max(mysize)], [df, df], 'k')
axchisq_mu_n.text(axchisq_mu_n.get_xlim()[1] * .6, 100, "expected mean", 
                  va="bottom", fontsize=20)
```

### Figure 2: [describe what you see and how it relates to the law of large numbers]

At low sample size the sample mean has a much higher variance, but as the sample size goes up the sample mean has a much smaller variance as predicted by the law of large number. The law of large numbers is as sample size increases the mean of the sample tends toward the true mean on the population

# Do it for the Normal distribution


```python
##your code here...
# fill dictionary in in a for loop. not pythonic, but easily readable
md['norm'] = {}
np.random.seed(56)
for n in mysize:
    md['norm'][n] = np.random.normal(mymean,scale = 10, size = n) #all distributions have mean=100
    #had to make scale/std bigger to not have all samples within 1 of 100
    
# save the means for plotting later    

md['norm']['means'] = {}
# prepare the figure
axnorm_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['norm']:
    if not type(nn) == str: # needed because one element is "mean"
        md['norm']['means'][nn] = md['norm'][nn].mean()
        #and plot it
        axnorm_mu_n.plot(nn, md['norm']['means'][nn], 'o')
        axnorm_mu_n.set_xlabel('sample size', fontsize=18)
        axnorm_mu_n.set_ylabel('sample mean', fontsize=18)
axnorm_mu_n.set_title('Normal', fontsize=18)
axnorm_mu_n.plot([min(mysize), max(mysize)], [df, df], 'k')
axnorm_mu_n.text(axnorm_mu_n.get_xlim()[1] * .6, 100, "expected mean", 
                  va="bottom", fontsize=20)
```

The graph follows the law of large numbers as sample mean aproaches population mean as sample size increases

```python
allmeans_n = list(md['norm']['means'].values())

pl.figure(figsize=(10, 10))
pl.hist(allmeans_n,bins=30)
pl.title('Normal Sample Means')
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

# Do it for the Binomial distribution

```python
##your code here...
# fill dictionary in in a for loop. not pythonic, but easily readable
md['bin'] = {}
np.random.seed(216)
for n in mysize:
    md['bin'][n] = np.random.binomial(2*mymean,0.5,size = n) #expected mean n*p => n = 2*mean
    
# save the means for plotting later    

md['bin']['means'] = {}
# prepare the figure
axbin_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['bin']:
    if not type(nn) == str: # needed because one element is "mean"
        md['bin']['means'][nn] = md['bin'][nn].mean()
        #and plot it
        axbin_mu_n.plot(nn, md['bin']['means'][nn], 'o')
        axbin_mu_n.set_xlabel('sample size', fontsize=18)
        axbin_mu_n.set_ylabel('sample mean', fontsize=18)
axbin_mu_n.set_title('Binomial', fontsize=18)
axbin_mu_n.plot([min(mysize), max(mysize)], [df, df], 'k')
axbin_mu_n.text(axbin_mu_n.get_xlim()[1] * .6, 100, "expected mean", 
                  va="bottom", fontsize=20)
```
